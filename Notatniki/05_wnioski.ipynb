# Wnioski

Przeprowadzona analiza danych mieszkaniowych obejmująca etapy wczytywania, czyszczenia, analizy opisowej oraz wizualizacji pozwoliła uzyskać pełny obraz struktury, dynamiki i zróżnicowania rynku nieruchomości w Polsce w okresie od sierpnia 2023 do czerwca 2024. Poniżej przedstawiono najważniejsze wnioski wynikające z całego procesu badawczego.

## 1. Jakość i struktura danych
Dane źródłowe okazały się spójne pod względem liczby kolumn i struktury, co umożliwiło ich bezproblemowe połączenie w jeden zbiór obejmujący ponad 195 tys. obserwacji. Występowanie powtarzających się identyfikatorów ofert było zjawiskiem naturalnym i oczekiwanym — wiele mieszkań pojawiało się w kolejnych miesiącach, co odzwierciedla realną dynamikę rynku.

Proces czyszczenia danych pozwolił na usunięcie jednej zmiennej oraz imputację pozostałych braków z wykorzystaniem metod dopasowanych do mechanizmów braków (MCAR/MAR). Zastosowanie IterativeImputer umożliwiło spójną i wielowymiarową imputację zmiennych odległościowych. Ostatecznie uzyskano kompletny, wysokiej jakości zbiór danych gotowy do dalszej analizy.

## 2. Charakterystyka rynku mieszkaniowego
Analiza opisowa wykazała, że rynek mieszkaniowy jest silnie zróżnicowany pod względem powierzchni, liczby pokoi, roku budowy oraz lokalizacji. Rozkłady zmiennych numerycznych były w większości prawoskośne, co wskazuje na obecność niewielkiej liczby mieszkań o bardzo dużych powierzchniach lub wysokich cenach.

Najsilniejszymi determinantami ceny okazały się powierzchnia oraz liczba pokoi, co potwierdziła zarówno korelacja, jak i wizualizacje. Zmienność cen rośnie wraz z wielkością mieszkań, co świadczy o większej heterogeniczności segmentu dużych apartamentów.

## 3. Znaczenie lokalizacji i typu zabudowy
Lokalizacja okazała się kluczowym czynnikiem wpływającym na cenę. Warszawa, Kraków i Gdańsk tworzą wyraźnie wyodrębniony segment premium, podczas gdy miasta średnie i mniejsze charakteryzują się niższymi i bardziej stabilnymi cenami. Testy ANOVA i Kruskal–Wallisa jednoznacznie potwierdziły istotne statystycznie różnice między miastami.

Typ zabudowy również odgrywa istotną rolę: apartamentowce są najdroższe, bloki najtańsze, a kamienice zajmują pozycję pośrednią. Różnice te wynikają zarówno z lokalizacji, jak i standardu budynków.

## 4. Trendy czasowe
Analiza zmian cen w czasie wykazała wyraźny trend wzrostowy od sierpnia 2023 do marca 2024, z największym przyspieszeniem w grudniu. Od kwietnia 2024 ceny ustabilizowały się, co może sugerować osiągnięcie równowagi rynkowej lub reakcję rynku na zmiany makroekonomiczne.

## 5. Dynamika ogłoszeń i trudność sprzedaży
Prawie połowa ogłoszeń (48,8%) pojawia się wielokrotnie, co wskazuje na trudności w sprzedaży części mieszkań. Najczęściej ogłoszenia powtarzają się 2–3 razy, ale istnieje grupa ofert „utkniętych” na rynku nawet przez 11 miesięcy.

Zmiany cen między powtórzeniami są częste — 28,5% ogłoszeń modyfikuje cenę, najczęściej obniżając ją o 4–6%. Wskazuje to na adaptację sprzedawców do warunków rynkowych. Zmienna `appearances` może stanowić wartościowy predyktor trudności sprzedaży.

## 6. Złożone zależności i heterogeniczność rynku
Zaawansowane wizualizacje (facet plots, hexbiny, violin plots, bubble charts, joint plots) ujawniły wielowymiarowe zależności między ceną, powierzchnią, liczbą pokoi, lokalizacją i typem zabudowy. Rynek mieszkaniowy jest silnie heterogeniczny, a kluczowe determinanty ceny działają w sposób nieliniowy i zależny od kontekstu.

## 7. Podsumowanie ogólne
Całość przeprowadzonej analizy wskazuje, że:
- cena mieszkań zależy przede wszystkim od lokalizacji, powierzchni i typu zabudowy,
- rynek charakteryzuje się dużą zmiennością i segmentacją,
- część mieszkań jest trudna do sprzedaży, co odzwierciedlają powtarzające się ogłoszenia,
- trendy cenowe w ostatnich miesiącach wskazują na dynamiczny wzrost, a następnie stabilizację,
- przygotowany zbiór danych jest kompletny i stanowi solidną podstawę do dalszych etapów pracy, w szczególności modelowania cen oraz analizy czynników wpływających na wartość nieruchomości.

Wnioski te tworzą spójny obraz rynku mieszkaniowego i stanowią fundament do budowy modeli predykcyjnych oraz dalszych badań analitycznych.
